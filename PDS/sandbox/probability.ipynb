# Adding two vectors `a` and `b`

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
iterations = 50

values = [0]
for i in range(iterations):
    if(np.random.random() >= 0.5):
        values.append(values[-1] + 1)
    else:
        values.append(max(0, values[-1] - 1))
```

```python
plt.title('Probability of a random walk')
plt.plot(values)
plt.xlabel('Iteration')
plt.ylabel('KM Travelled')
plt.show()
```
