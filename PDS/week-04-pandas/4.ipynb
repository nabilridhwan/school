```python
import numpy as np
import pandas as pd
```

```python
myseries = pd.Series(np.random.rand(1000));
print(myseries)
```

```python
colA = np.random.randint(10, 100, 1000)
colB = np.zeros(1000, dtype='u8')
colC = np.arange(1000)

mydf = pd.DataFrame(np.array([colA, colB, colC]).T)
print(mydf)
```

```python
print(myseries.shape)
print(myseries.index)
print(myseries.values)
print(myseries.count())
print(myseries.describe())
```

```python
print(mydf.shape)
print(mydf.index)
print(mydf.values)
print(mydf.count())
print(mydf.describe())
```
