```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_rainfall = pd.read_csv("./rainfall-monthly-total.csv")
# Change datatype of month column to datetime
df_rainfall["month"] = pd.to_datetime(df_rainfall["month"])

df_1982 = df_rainfall[df_rainfall["month"].dt.year == 1982]

# Generate datetime object for each month
months = pd.date_range(start="1982-01-01", end="1982-12-31", freq="MS")

# Set the size of plot
plt.figure(figsize=(12, 5))
plt.plot(df_1982["month"], df_1982["total_rainfall"])
plt.xticks(months)
plt.ylabel("total_rainfall")
plt.legend()
plt.xlabel("month")
plt.show()
```
