```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_rainfall = pd.read_csv("./rainfall-monthly-total-withduplicates.csv")
df_rainfall.index = df_rainfall["month"]

data = df_rainfall.drop_duplicates().sort_values(by="total_rainfall", ascending=False).iloc[-12:, :]

plt.figure(figsize=(10, 5))
plt.bar(data["month"], data["total_rainfall"])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.legend()
plt.show()
```
