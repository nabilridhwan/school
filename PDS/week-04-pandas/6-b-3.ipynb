```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_rainfall = pd.read_csv("./rainfall-monthly-total.csv")
df_rainfall.index = df_rainfall["month"]

morethan300mm = df_rainfall[df_rainfall["total_rainfall"] > 300].sort_values("total_rainfall", ascending=True)

morethan300mm_top_twelve = morethan300mm.iloc[-12:,:]

plt.figure(figsize=(10, 5))
plt.bar(morethan300mm_top_twelve.index, morethan300mm_top_twelve["total_rainfall"])
plt.xlabel("Month")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
